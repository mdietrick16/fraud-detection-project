# Data Exploration

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

#### Load Dataset

```python
file_name = '../data/insuranceFraud.csv'
data = pd.read_csv(file_name)
```

#### Inspect Data

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

#### Check for Missing Values

```python
data.isnull().sum()
```

#### Distribution of Numerical Features

```python
data.hist(bins=50, figsize=(20, 15))
plt.show()
```

#### Distribution of Categorical Features

```python
categorical_columns = data.select_dtypes(include=['object']).columns

fig, axes = plt.subplots(5, 5, figsize=(20, 30))

axes = axes.flatten()


for i, column in enumerate(categorical_columns):
    sns.countplot(y=data[column], ax=axes[i])
    axes[i].set_title(column)

for j in range(i+1, len(axes)): 
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

for column in categorical_columns: 
    unique_values = data[column].unique() 
    print(f'{column}:\n{unique_values[:10]}\n')
```

```python
data[categorical_columns].describe()
```

#### Inspect Outliers

```python
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
numerical_columns = numerical_columns.drop('_c39')

num_columns = len(numerical_columns) 
num_rows = (num_columns + 1) // 2

fig, axes = plt.subplots(5, 4, figsize=(20, 15)) 

axes = axes.flatten() 

for i, column in enumerate(numerical_columns): 
    sns.boxplot(x=data[column], ax=axes[i]) 
    axes[i].set_title(column)

for j in range(i+1, len(axes)): 
    fig.delaxes(axes[j]) 

plt.tight_layout() 
plt.show()
```

#### Correlation Analysis

```python
# Set the figure size for the correlation heatmap
plt.figure(figsize=(18, 15))

# Calculate the correlation matrix for numerical columns in the data
corr = data.select_dtypes(include=['number']).corr()

# Plot the heatmap of the correlation matrix with annotations and specified formatting
sns.heatmap(data=corr, annot=True, fmt='.1g', linewidth=2)

# Display the plot
plt.show()
```

```python
# Checking if any of our numerical features are strongly related to each other (aka, correlated).
# If two features are super similar, we might not need both. 
# This just gives us a sense of whatâ€™s related in our data.
plt.figure(figsize = (18,15))
corr = data.select_dtypes(include=['number']).corr()
mask = np.triu(np.ones_like(corr,dtype =bool))

sns.heatmap(data =corr, mask = mask, annot = True, fmt ='.2g', linewidth =1)
plt.show
```
